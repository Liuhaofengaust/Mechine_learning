## 回归
* 线性回归
* Lasso回归-利用L1正则化：添加一个与所有权重绝对值的和成比例的元素
* ridge回归-利用L2正则化：添加一个与所有权重平方和成比例的元素
* 逻辑回归

## 1-线性回归

```python
import numpy as np
from sklearn import datasets, metrics, model_selection as modsel, linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

### 载入数据集

```python
boston = datasets.load_boston()
```

```python
print(dir(boston))
print(boston.data.shape)
```

* 数据集包含506个数据点，每个数据点包含13个特征

### 载入模型

```python
linreg = linear_model.LinearRegression()

x_train, x_test, y_train, y_test = modsel.train_test_split(
    boston.data, boston.target, test_size=0.1, random_state=42)
```

### 训练模型

```python
linreg.fit(x_train, y_train)
```

### 评估模型

```python
print('训练均方误差：', metrics.mean_squared_error(y_train, linreg.predict(x_train)))
print('训练的R方值：', linreg.score(x_train, y_train))

y_pred = linreg.predict(x_test)
print('测试数据上的均方误差',metrics.mean_squared_error(y_test, y_pred))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y_test, linewidth=3, label='ground truth')
plt.plot(y_pred, linewidth=3, label='predicted')
plt.legend(loc='best')
plt.xlabel('test data points')
plt.ylabel('target value')
```

```python
# 更专业（直观）的绘图
plt.plot(y_test, y_pred, 'o')
plt.plot([-10, 60], [-10, 60], 'k--')
plt.axis([-10, 60, -10, 60])
plt.xlabel('ground truth')
plt.ylabel('predicted')
scorestr = r'R$^2$ = %.3f' %linreg.score(x_test, y_test)
errstr = 'MSE = %.3f' %metrics.mean_squared_error(y_test, y_pred)
plt.text(-5, 50, scorestr, fontsize=12)
plt.text(-5, 45, errstr, fontsize=12)
```

* 对角线表示y_test和y_pred相等
* r^2表明我们可以解释数据的76%的离散度

## 2-Lasso回归

```python
lassoreg = linear_model.Lasso()

lassoreg.fit(x_train, y_train)

print('训练均方误差：', metrics.mean_squared_error(y_train, lassoreg.predict(x_train)))
print('训练的R方值：', lassoreg.score(x_train, y_train))

y_pred = lassoreg.predict(x_test)
print('测试数据上的均方误差',metrics.mean_squared_error(y_test, y_pred))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y_test, linewidth=3, label='ground truth')
plt.plot(y_pred, linewidth=3, label='predicted')
plt.legend(loc='best')
plt.xlabel('test data points')
plt.ylabel('target value')
```

```python
# 更专业（直观）的绘图
plt.plot(y_test, y_pred, 'o')
plt.plot([-10, 60], [-10, 60], 'k--')
plt.axis([-10, 60, -10, 60])
plt.xlabel('ground truth')
plt.ylabel('predicted')
scorestr = r'R$^2$ = %.3f' %lassoreg.score(x_test, y_test)
errstr = 'MSE = %.3f' %metrics.mean_squared_error(y_test, y_pred)
plt.text(-5, 50, scorestr, fontsize=12)
plt.text(-5, 45, errstr, fontsize=12)
```
